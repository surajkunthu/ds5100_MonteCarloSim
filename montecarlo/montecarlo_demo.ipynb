```python
from montecarlo import Die, Game, Analyzer
import matplotlib.pyplot as plt
```

# Scenario 1: 2-headed coin
1. Create one fair coin (with faces H and T) and one unfair coin. For the unfair coin, give one of the faces a weight of 5 and the others 1.

```python
fair = Die(faces = ["H", "T"])
unfair = Die(faces = ["H", "T"])
fair.show_die(), unfair.show_die()
```

```python
unfair.change_weight("H", 5)
unfair.show_die()
```

2. Play a game of 1000 flips of three coins with all fair dice.

```python
game_fair = Game(object_list= [fair, fair, fair])
game_fair.play(1000)
```

3. Play a game of 1000 flips with two unfair dice and one fair die.

```python
game_mixed = Game(object_list= [fair, unfair, unfair])
game_mixed.play(1000)
```

4. For each game, use an Analyzer object to determine the relative frequency of jackpots – getting either all Hs or all Ts.

```python
analyze_fair = Analyzer(game_fair)
fair_freq = analyze_fair.jackpot()
print(fair_freq)
```

```python
analyzer_mixed = Analyzer(game_mixed)
mixed_freq = analyzer_mixed.jackpot()
print(mixed_freq)
```

5. Compute relative frequency as the number of jackpots over the total number of rolls.

- Relative frequency for all Fair coins is $255/1000$ = $25\%$
- Relative frequency for Two unfair coins and one fair coin is $293/1000$ = $29.3\%$

6. Show your results, comparing the two relative frequencies, in a simple bar chart.

```python
plt.bar(x=["Fair Coins Jackpot Frequency", "Mixed Coins Jackpot Frequency"], height=[fair_freq, mixed_freq]);
plt.ylim(0, 1000);
```

# Scenario 2: 6-sided die
1. Create a fair die and two unfair dice, all of six sides with the faces 1 through 6. One unfair die (Type 1) will weight 6 five times more than the others (i.e. it has weight of 5 and the others a weight of 1 each). The other unfair die (Type 2) will weight 1 five times more than the others.

```python
fair_die = Die(faces= [1,2,3,4,5,6])
unfair_type1 = Die(faces=[1,2,3,4,5,6])
unfair_type2 = Die(faces=[1,2,3,4,5,6])

unfair_type1.change_weight(6, 5)
unfair_type2.change_weight(1, 5)

fair_die.show_die(), unfair_type1.show_die(), unfair_type2.show_die()
```

2. Play a game of 10000 rolls with 5 fair dice.

```python
fair_5 = Game(object_list = [fair_die,fair_die,fair_die,fair_die,fair_die])
fair_5.play(10000)
```

3. Play a game of 10000 rolls with 2 unfair dice of type 1, 1 unfair die of type 2, and the rest fair dice.

```python
unfair_mixed = Game(object_list=[unfair_type1, unfair_type1, unfair_type2, fair_die, fair_die])
unfair_mixed.play(10000)
```

4. For each game, use an Analyzer object to determine the relative frequency of jackpots and show your results, comparing the two relative frequencies, in a simple bar chart.

```python
fair_5_jackpot = Analyzer(fair_5)
fair_5_jackpotScore = fair_5_jackpot.jackpot()
fair_5_jackpotScore
```

```python
unfair_mixed_jackpot  = Analyzer(unfair_mixed)
unfair_mixed_jackpotScore = unfair_mixed_jackpot.jackpot()
unfair_mixed_jackpotScore
```

The relative frequencies:
- 5 Fair Die: $6/10000$ = $0.06\%$
- Mixed Unfair Die (2 unfair dice of Type 1, 1 unfair dice of Type 2, 2 fair dice): $8/10000$ = $0.08\%$

```python
plt.bar(x=["5 Fair Dice Jackpot Frequency", "Unfair Mixed Dice Frequency"],
            height=[fair_5_jackpotScore, unfair_mixed_jackpotScore]);
#plt.ylim(0, 10000);
```

# Scenario 3: Roman Alphabet
1. Create a "die" of letters from a to z with weights based on their frequency of usage. See Appendix for these weights.

```python
roman_alphabet = Die(faces=["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"])
roman_faces = ["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]
roman_weights = [8.4966,2.0720,4.5388,3.3844,11.1607,1.8121,2.4705,3.0034,7.5448,0.1965,1.1016,5.4893,3.0129,6.6544,7.1635,3.1671,0.1962,7.5809,5.7351,6.9509,3.6308,1.0074,1.2899,0.2902,1.7779,0.2722]
```

```python
i = 0
while i < len(roman_faces):
    roman_alphabet.change_weight(roman_faces[i], roman_weights[i])
    i += 1
```

```python
roman_alphabet.show_die()
```

2. Play a game involving 5 of these dice with 1000 rolls.

```python
roman_game = Game(object_list = [roman_alphabet, roman_alphabet, roman_alphabet, roman_alphabet, roman_alphabet])
roman_game.play(1000)
```
